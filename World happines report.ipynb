```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(10,7)
plt.rcParams['figure.facecolor']='#FFE5B4'
```

```python
data=pd.read_csv(r'D:\work\data sets\World happines report/world-happiness-report-2021.csv')
data
```

```python
data.keys()
```

```python
data_columns=['Country name', 'Regional indicator', 'Ladder score','Logged GDP per capita','Social support','Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
```

```python
data=data[data_columns].copy()
```

```python
data.head()
```

```python
happy_df=data.rename({'Countryname':'Country_name','Regionalindicator':'Regional_indicator','Ladderscore':'Ladder_score','LoggedGDPpercapita':'Logged_GDP_per_capita'},inplace=False)
```

```python
happy_df.head()
```

```python
happy_df.isnull().sum()
```

```python
#plot between happiness and GDP

plt.rcParams['figure.figsize']=(15,7)
plt.title('plot between happiness score and GDP')
sns.scatterplot(x=happy_df.Ladder score,y=happy_df.Logged GDP per capita,hue=happy_df.Regional indicator,s=200)
plt.legend(loc = 'upper left',fontsize='10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per capita ')
```

```python
gdp_region= data.groupby('Regional indicator')['Logged GDP per capita'].sum()
gdp_region
```

```python
gdp_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP BY REGION')
plt.ylabel('')
```

```python
# total countries

total_country=data.groupby('Regional indicator')[['Country name']].count()
total_country
```

```python
# correlation map

cor=data.corr(method="pearson")
f,ax=plt.subplots(figsize=(10,5))
sns.heatmap(cor,mask=np.zeros_like(cor,dtype=np.bool),cmap="Blues",square=True,ax=ax);
```

```python
# correption in diffrent regions

corruption =data.groupby('Regional indicator')[['Perceptions of corruption']].mean()
corruption
```

```python
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of correcption in various regions')
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha='right')
plt.bar(corruption.index,corruption.Perceptionsofcorruption)
```
