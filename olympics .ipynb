```python
#top 10 countries participating

top_10=data.Team.value_counts().sort_values(ascending=False).head(10)
top_10

```

```python
# plot the graph

plt.figure(figsize=(12,6))
plt.title('top 10 participants')
sns.barplot(x=top_10.index,y=top_10,palette='Set2');
```

```python
# age distribution of the participants

plt.figure(figsize=(12,6))
plt.title("age distibution of athelets")
plt.xlabel("Age")
plt.ylabel("no of participants")
plt.hist(data.Age,bins=np.arange(10,80,2),color='orange',edgecolor='white');
```

```python
# male and female participants
gender_counts=data.Sex.value_counts()
gender_counts
```

```python
# pie plot for male and female athletes
plt.figure(figsize=(12,6))
plt.title(" gender distribution")
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150);
```

```python
# total number of medals
total_medal=data.Medal.value_counts
total_medal
```

```python
# total no of female athelates in each olympics
female_participants=data[(data.Sex=='F')&(data.Season=='Winter')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants
```

```python
female_participants=data[(data.Sex=='F')&(data.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants
```

```python
women_olympics=data[(data.Sex=='F')&(data.Season=='Winter')]
```

```python
sns.set(style="darkgrid");
plt.figure(figsize=(20,10));
sns.countplot(x='Year',data=women_olympics,palette='Spectral');
plt.title('women participation');
```

```python
# ploting line graph
part=women_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,"F"].plot()
plt.title('plot of female athelets over a time');
```

```python
# gold medal athelets
goldmedals=data[(data.Medal=='Gold')]
goldmedals.head()
```

```python
# take the values that are diffrent from NaN
goldmedals=goldmedals[np.isfinite(goldmedals['Age'])]
goldmedals
```

```python
#gold medal beyond 60
goldmedals['ID'][goldmedals['Age']>60].count()
```

```python
# in this sports
sporting_event=goldmedals['Sport'][goldmedals['Age']>60]
sporting_event
```

```python
#plot for sporting_event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('gold medals for athelets age over 60 years');
```

```python
# gold medal for each country
data.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
# rio olympic
max_year=data.Year.max()
max_year

team_names=data[(data.Year==max_year)&(data.Medal=='Gold')].Team
team_names.value_counts().head(10)
```

```python
# bar plot
plt.figure(figsize=(12,6))
sns.barplot(x=team_names.value_counts().head(20),y=team_names.value_counts().head(20).index)
plt.title('countrywise medal for year 2016(rio olympics)');
```
